```python
x = range(1, 100)
x2 = np.array(x)
```

```python
y =[]
for z in x:
    y.append(np.log(z))
y2 = np.array(y)
```

```python
y3 = y2*y2
x3 = x2*x2
```

```python

figure, graficos = plt.subplots(nrows=3, ncols=2)
graficos[0][0].plot(x2, x3, 'r')
graficos[0][1].plot(y2, y3, 'y')
graficos[1][1].plot(y3)
graficos[1][0].plot(x3)
graficos[2][0].plot(x3*x2)
graficos[2][1].plot(y3*y2)
```

```python
figura = plt.figure()
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(y2, y2**2, 'red', label='X al cuadrado')
graficos.plot(x2, x2**2, 'orange', label='Y**2 vs Y**3')
graficos.legend(loc=0)
```

```python
fig = plt.figure()
graf = fig.add_axes([0,0,1,1])
```

```python
figu = plt.figure()
grafi = figu.add_axes([0,0, 1, 1])
grafi.plot(x2, x2**2, 'red', linewidth=5, linestyle=':')
```
