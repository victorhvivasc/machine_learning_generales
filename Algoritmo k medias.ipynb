<h1>Algoritmo k medias

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
datos = make_blobs(n_samples=200, n_features=2, centers=4)
```

```python
plt.scatter(datos[0][:,0], datos[0][:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmedias = KMeans(n_clusters=4)
```

```python
kmedias.fit(datos[0])
```

<h3>En este modelo no hay predicciones, se puede consultar la posicion central de los clubster

```python
dato=kmedias.cluster_centers_
```

```python
x1, y1 = [], []
for i in dato:
    x1.append(i[0])
    y1.append(i[1])  
```

```python
plt.scatter(x1, y1)
```
